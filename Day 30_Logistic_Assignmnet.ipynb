```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Importing data form sklearn
from sklearn.datasets import load_digits
digits = load_digits()
digits.data
```

```python
#Loading into dataframe
data = pd.DataFrame(digits.data)
data.head()
```

```python
#Splitting the data into training and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(digits.data,digits.target,test_size=0.25,random_state=41)
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler
s_x = StandardScaler()
x_train = s_x.fit_transform(xtrain)
x_test = s_x.transform(xtest)
```

```python
#Building model
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train,ytrain)
#Predictions
ypred = log_reg.predict(x_test)
```

```python
#Evaluation Metrics
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cmd = pd.DataFrame(cm,index=digits.target_names)
print("Confusion matrix:\n",cmd)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
print("Accuracy :",np.round(accuracy_score(ytest,ypred),2))
```

```python
#Classification report
from sklearn.metrics import classification_report
print("Classification report as follows\n:")
print(classification_report(ytest,ypred))
```
