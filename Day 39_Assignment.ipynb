```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# For market basket analysis
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Loading dataset
df = pd.read_csv("C:\Sharath\DATA SETS\ASSOCIATION RULES\\BreadBasket.csv")
print(df.shape)  #(21293,4)
df.nunique() # Transaction = 9531 , Item = 95
```

```python
#Data
data = df[['Transaction','Item']]
data['Item'].unique()
```

```python
#Grouping the data
basket = data.groupby(['Transaction','Item'])['Item'].count().unstack().fillna(0)
basket
```

```python
#Replacing the count with 1 or 0
basket_model = basket.applymap(lambda x:1 if x>0 else 0)
basket_model
```

```python
#Applying apriori algorithm
popular_sets = apriori(basket_model,min_support=0.03,use_colnames=True)
popular_sets
```

```python
#Rules
rules = association_rules(popular_sets,metric='lift',min_threshold=1)
rules
```
