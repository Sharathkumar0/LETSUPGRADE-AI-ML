```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Importing and Loading data from sklearn
from sklearn.datasets import load_iris
iris = load_iris()
iris.feature_names
iris.target_names
```

```python
#Viewing the data in the dataframe
data = pd.DataFrame(iris.data)
data.columns = iris.feature_names
data.head()
```

```python
#Splitting the data for training and testing
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(iris.data,iris.target,stratify = iris.target,
                                             test_size=0.25,random_state=1)
```

```python
#Model building
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(xtrain,ytrain)
```

```python
#Predictions
ypred = classifier.predict(xtest)
```

```python
#Accuracy 
from sklearn.metrics import accuracy_score
print("Accuracy score :",np.round(accuracy_score(ytest,ypred),2))
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cmd = pd.DataFrame(cm,index=iris.target_names,columns=iris.target_names)
print("Confusion Matrix :\n ",cmd)
```

```python
#Classification Report 
from sklearn.metrics import classification_report
print("Classification Report :\n")
print(classification_report(ytest, ypred))
```
